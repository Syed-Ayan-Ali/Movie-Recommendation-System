### There are 3 null values in 'keywords' so we need to remove them

```python
movies.dropna(inplace=True)
```

```python
genres = movies.iloc[0].genres
genres
```

```python
import ast

def convert(genres):
    genre_list = []
    for genre in ast.literal_eval(genres):
        genre_list.append(genre['name'])
    return genre_list

convert(genres)
    
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
cast = movies.iloc[0].cast
cast
```

```python
def convert_cast(cast):
    count = 0
    top_3_cast = []
    for c in ast.literal_eval(cast):
        top_3_cast.append(c['character'])
        count += 1
        if (count == 3):
            break
            
    return top_3_cast    

convert_cast(cast)
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
crew = movies.iloc[0].crew
crew
```

```python
def get_director(crew):
    directors = []
    for c in ast.literal_eval(crew):
        if (c['job'] == "Director"):
            directors.append(c['name'])
            break
            
    return directors
```

```python
movies['crew'] = movies['crew'].apply(get_director)
```

```python
movies.head()
```

### Convert overview to a list

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# We need to combine the words with spaces to avoid information loss
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "" ) for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "" ) for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "" ) for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "" ) for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] +  movies['keywords'] +  movies['cast'] +  movies['crew'] 
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df
```

```python
## convert tag to text
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
import nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[1]
```

```python
cv.get_feature_names_out()
```

```python
## Calculate cosine distance between movies
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
 similarity[0]
    
## This is the similarity of this movie with all the other movies.
```

```python
enum_similarity_vector = list(enumerate(similarity[0]))
# preserves the index and the similarity of the movie with respect to other movies
```

```python
# sort to get the top 5 closest movies to the original movie
sorted(enum_similarity_vector, reverse=True, key = lambda x:x[1])
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    enum_similarity_vector = list(enumerate(similarity[movie_index]))
    sorted_similarity_vector = sorted(enum_similarity_vector, reverse=True, key = lambda x:x[1])[1:6]
    
    
    movies = []
    for i in range(5):
        movies.append(new_df.title[sorted_similarity_vector[i][0]])
    
    return movies

def available(movie):
    for i in new_df['title']:
        if i == movie:
            print(movie)
            return True
        
    
    return False

available('Avatar')
recommend('The Beautiful Mind')
```
